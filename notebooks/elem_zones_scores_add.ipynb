```python
elem_zones_order = elem_zones_first.sort_values(['SCHOOL_NAM'])
elem_zones_order.head(25)
```

```python
elem_scores_order = elem_scores.sort_values(['SCHOOL'])
elem_scores_order.head(26)
```

```python
elem_zones_order = elem_zones_order.reset_index()
elem_zones_order.head(10)
```

```python
elem_scores_order = elem_scores_order.reset_index()
elem_scores_order.head(10)
```

```python
elem_zones_scores = elem_zones_order.merge(elem_scores_order, how='outer', left_index=True, right_index=True)
```

```python
elem_zones_scores.tail()
```

```python
elem_zones_scores.shape
```

```python
elem_zones_scores = elem_zones_scores[['SCHOOL_NAM',
                                           'ACHIEVEMENT_SCORE',
                                           'GROWTH_SCORE',
                                           'CHRONICALLY_OUT_OF_SCHOOL_SCORE',
                                           'geometry']]
elem_zones_scores.head()
```

```python
elem_zones_scores = elem_zones_scores.rename(columns={'SCHOOL_NAM': 'school',
                                                          'ACHIEVEMENT_SCORE': 'achievement_score',
                                                          'GROWTH_SCORE': 'growth_score',
                                                          'CHRONICALLY_OUT_OF_SCHOOL_SCORE': 'chronically_out_of_school_score'})

elem_zones_scores.head()
```

```python
f, ax = plt.subplots(1, figsize=(15, 15))
elem_zones_scores.plot(column = 'school', edgecolor = 'black', ax = ax, legend = True)
plt.show()
```

```python
elem_zones_scores_gdf = gpd.GeoDataFrame(elem_zones_scores, geometry='geometry')
elem_zones_scores_gdf.to_file('../data/final_data/elem_zones_scores_gdf.geojson', driver = 'GeoJSON')
```
