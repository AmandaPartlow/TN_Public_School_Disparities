```python
middle_zones_order = middle_zones_first.sort_values(['SCHOOL_NAM'])
middle_zones_order.head(25)
```

```python
middle_scores_order = middle_scores.sort_values(['SCHOOL'])
middle_scores_order.head(26)
```

```python
middle_zones_order = middle_zones_order.reset_index()
middle_zones_order.head(26)
```

```python
middle_scores_order = middle_scores_order.reset_index()
middle_scores_order.head(26)
```

```python
middle_zones_scores = middle_zones_order.merge(middle_scores_order, how='outer', left_index=True, right_index=True)
```

```python
middle_zones_scores.head()
```

```python
middle_zones_scores.shape
```

```python
middle_zones_scores = middle_zones_scores[['SCHOOL_NAM',
                                           'ACHIEVEMENT_SCORE',
                                           'GROWTH_SCORE',
                                           'CHRONICALLY_OUT_OF_SCHOOL_SCORE',
                                           'geometry']]
middle_zones_scores.head()
```

```python
middle_zones_scores = middle_zones_scores.rename(columns={'SCHOOL_NAM': 'school',
                                                          'ACHIEVEMENT_SCORE': 'achievement_score',
                                                          'GROWTH_SCORE': 'growth_score',
                                                          'CHRONICALLY_OUT_OF_SCHOOL_SCORE': 'chronically_out_of_school_score'})

middle_zones_scores.head()
```
