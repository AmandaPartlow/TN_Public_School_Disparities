```python
elem_zones = gpd.read_file('../data/mnps/MNPS_ESZone_2021.shp')
```

```python
elem_zones.head()
```

```python
elem_zones.crs = 'EPSG:4326'
```

```python
elem_zones = elem_zones.to_crs('EPSG:4326')
```

```python
f, ax = plt.subplots(1, figsize=(15, 15))
elem_zones.plot(column = 'SCHOOL_NAM', edgecolor = 'black', ax = ax, legend = True)
plt.show()
```

```python
elem_zones.shape
```
