```python
import pandas as pd
import geopandas as gpd
import requests
from shapely.geometry import Point
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
per_cap = gpd.read_file('../data/final_data/davidson_per_cap.geojson')
race = gpd.read_file('../data/final_data/davidson_race.geojson')
income = gpd.read_file('../data/final_data/davidson_income.geojson')
elem = gpd.read_file('../data/final_data/elem_zones_scores_gdf.geojson')
middle = gpd.read_file('../data/final_data/middle_zones_scores_gdf.geojson')
high = gpd.read_file('../data/final_data/high_zones_scores_gdf.geojson')
```

```python
elem_middle = gpd.sjoin(elem, middle, op = 'intersects')
elem_middle.head(2)
```

```python
elem_middle.shape
```

```python
elem_middle = elem_middle.rename(columns={'school_left': 'elem_school', 
                                          'achievement_score_left': 'elem_achievement_score',
                                          'growth_score_left': 'elem_growth_score',
                                          'chronically_out_of_school_score_left': 'elem_chronically_out_of_school_score',
                                          'school_right': 'middle_school', 
                                          'achievement_score_right': 'middle_achievement_score',
                                          'growth_score_right': 'middle_growth_score',
                                          'chronically_out_of_school_score_right': 'middle_chronically_out_of_school_score'})

elem_middle = elem_middle[elem_middle.columns.drop('index_right')]

elem_middle.head()
```

```python
schools = gpd.sjoin(elem_middle, high, op = 'intersects')
schools.head()
```

```python
schools = schools.rename(columns={'school': 'high_school', 
                                          'achievement_score': 'high_achievement_score',
                                          'growth_score': 'high_growth_score',
                                          'chronically_out_of_school_score': 'high_chronically_out_of_school_score'})

schools = schools[schools.columns.drop('index_right')]

schools.head()
```

```python
schools.shape
```
